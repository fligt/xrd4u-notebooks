## Bragg formula intermezzo 

After some algebra it initially seemed that the theta angle multiplication factor would be simply the ratio for the two wavelengths. However for typical crystal mono-chromators  there seems to be an additional Lorentz-Polarization correction factor

https://dictionary.iucr.org/Lorentz%E2%80%93polarization_correction


Here is the Bragg formula: $$ n \lambda = \frac{2d}{\sin \theta} $$
This implies that the ratio $\theta' / \theta$ 

For an derivation of Bragg's formula see [wikipedia](https://en.wikipedia.org/wiki/Bragg%27s_law). 

![](https://upload.wikimedia.org/wikipedia/commons/0/0a/Bragg%27s_law.svg)

```python
import scipy.constants as sc
import numpy as np
```

```python
wavelength = 9.510908054342373e-11 # reported in json 
```

Need to convert this to keV... Ehm $f = c / \lambda$ and $E = h \times f$. Combining both gives: $E = \frac{h \times c}{\lambda}$. 

```python
E_joule = sc.h * sc.c / wavelength
E_joule
```

Next we need to convert the energy from Joules to keV... 

1 eV = 1.602176634e-19 J 

```python
E_keV = 0.001 * E_joule / sc.e
E_keV
```

```python
lambda_0 = 1
lambda_1 = 2
theta_factor = np.arcsin(0.5 * lambda_1 / lambda_0) / (sc.pi / 4)
theta_factor
```
